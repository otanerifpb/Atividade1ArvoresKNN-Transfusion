Passo 01 - Carregar a base TRANSFUSION. Por uma URL

```python
import pandas as pd

url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/transfusion.data"

df = pd.read_csv(url, header=None)
columns = len(df.columns)

df
```

Passo 02 - A coluna 0 (zero) é o TARGET. Colunas 1 a 4 Features(X)

```python
y = df[0]


X = df.loc[:,1:columns-1]

print(y)
print()
X
```

Passo 03 - Observe que a magnitude dos atributos é bem diferente, então cada atributo deve ser normalizado (código no mini tutorial de pandas).

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

#df[1] = MinMaxScaler().fit_transform(np.array(df[1]).reshape(-1,1))
#df[2] = MinMaxScaler().fit_transform(np.array(df[2]).reshape(-1,1))
#df[3] = MinMaxScaler().fit_transform(np.array(df[3]).reshape(-1,1))
#df[4] = MinMaxScaler().fit_transform(np.array(df[4]).reshape(-1,1))

df[df.columns[1:]] = MinMaxScaler().fit_transform(df[df.columns[1:]])

df
```

Passo 04 - Separar os dados de treinamento e testes

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = None, stratify = y)
```

Passo 05 - Treinar 2 árvores de decisão com características diferentes

```python
from sklearn.tree import DecisionTreeClassifier

# Árvore de Decisão com entropy
treino1 = DecisionTreeClassifier(criterion = "entropy")
#treino1 = DecisionTreeClassifier(criterion = "gini")
# Árvore de Decisão com profundidade limitada
treino2 = DecisionTreeClassifier(max_depth = 3)
#treino2 = DecisionTreeClassifier(max_depth = 5)

# Treinameto dos dados com Árvore de Decesão
treino1.fit(X_train, y_train)
treino2.fit(X_train, y_train)
```

Passo 06 - Treinar 2 kNN com características diferentes

```python
from sklearn.neighbors import KNeighborsClassifier

# K-NN com 7 vizinhos, metrica 'euclidoen'
treino3 = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean', algorithm = 'brute')
# K-NN com 5 vizinhos, metrica 'manhattan'
treino4 = KNeighborsClassifier(n_neighbors = 7, metric = 'manhattan', algorithm = 'brute')

# Treinameto dos dados com Árvore de Decesão
treino3.fit(X_train, y_train)
treino4.fit(X_train, y_train)
```

Passo 07 - Exibir acurácia dos testes

```python
from sklearn import metrics

print('*'*13 + 'Acurácia Árvore Decisão' + '*'*15)
result1 = treino1.predict(X_test)
acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1 * 100)
print("Árvore Decisão Profundidade livre: " + ' '*13 + "{}%".format(show1))

result2 = treino2.predict(X_test)
acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2 * 100)
print("Árvore Decisão Profundidade 3: " + ' '*17 + "{}%".format(show1))

print()
print('*'*19 + 'Acurácia K-NN' + '*'*19)
result3 = treino3.predict(X_test)
acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3 * 100)
print("Protocolo K-NN 7 Neighbord Metric 'eucledean': " + ' ' + "{}%".format(show3))

result4 = treino4.predict(X_test)
acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4 * 100)
print("Protocolo K-NN 7 Neighbord  Metric 'manhattan': " + "{}%".format(show3))
```
